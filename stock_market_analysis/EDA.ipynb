```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hist_nvda = pd.read_csv("data/NVDA.csv", index_col='Date', parse_dates=True)
hist_tsla = pd.read_csv("data/TSLA.csv", index_col='Date', parse_dates=True)
hist_jpm = pd.read_csv("data/JPM.csv", index_col='Date', parse_dates=True)
# Concatenate the closing prices into a DataFrame
df = pd.concat([hist_nvda['Close'], hist_tsla['Close'], hist_jpm['Close']], axis=1)
df.columns = ['NVDA', 'TSLA', 'JPM']
```

```python
plt.figure(figsize=(12,6))
for column in df.columns:
    plt.plot(df.index, df[column])
    plt.title('NVDA, TSLA, JPM Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.grid(True)
    legend_labels = ['NVDA', 'TSLA', 'JPM']
# Create Custom Colors 
colors = {'NVDA':'green', 'TSLA':'white', 'JPM':'brown'}
legend_handles = [plt.Line2D([0], [0], color=colors[label], linewidth=2, linestyle='-') for label in legend_labels]
plt.legend(legend_handles, legend_labels, loc='upper left')
```

**NVDA (NVIDIA Corporation):**
- NVIDIA's stock prices have generally trended upwards since 2020, with periodic fluctuations reflecting market dynamics and company performance.

**JPM (JPMorgan Chase & Co.):**
- JPMorgan's stock prices have shown relative stability, influenced by factors in the financial sector and economic conditions.

**TSLA (Tesla):**
- Tesla's stock prices have been volatile, experiencing significant growth driven by factors such as market sentiment, company developments, and demand for electric vehicles.

Analyzing these closing prices can offer insights into the historical performance of these companies, highlighting trends and fluctuations in the stock market over the specified period.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df contains the stock price data for NVDA, JPM, and TSLA

# Calculate moving averages
df['NVDA_50_MA'] = df['NVDA'].rolling(window=50).mean()
df['NVDA_200_MA'] = df['NVDA'].rolling(window=200).mean()

df['JPM_50_MA'] = df['JPM'].rolling(window=50).mean()
df['JPM_200_MA'] = df['JPM'].rolling(window=200).mean()

df['TSLA_50_MA'] = df['TSLA'].rolling(window=50).mean()
df['TSLA_200_MA'] = df['TSLA'].rolling(window=200).mean()
```

```python
# Plot moving averages for NVDA
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['NVDA'], label='NVDA')
plt.plot(df.index, df['NVDA_50_MA'], label='NVDA 50-Day MA')
plt.plot(df.index, df['NVDA_200_MA'], label='NVDA 200-Day MA')
plt.legend()
plt.title('NVDA Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Plot moving averages for TSLA
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['TSLA'], label='TSLA')
plt.plot(df.index, df['TSLA_50_MA'], label='TSLA 50-Day MA')
plt.plot(df.index, df['TSLA_200_MA'], label='TSLA 200-Day MA')
plt.legend()
plt.title('TSLA Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Plot moving averages for JPM 
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['JPM'], label='JPM')
plt.plot(df.index, df['JPM_50_MA'], label='JPM 50-Day MA')
plt.plot(df.index, df['JPM_200_MA'], label='JPM 200-Day MA')
plt.legend()
plt.title('JPM Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
